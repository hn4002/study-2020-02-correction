```python
import pandas
import matplotlib.pyplot as plt
```

```python
rating='EPS Rating'
ratingS='EPSRating'
df1 = pandas.read_csv("results/stats-H1-2019-{}.csv".format(ratingS), names=['X', rating, 'H1-2009-' + ratingS], skiprows=[0])
df1.head()
```

```python
ratingC='Composite Rating'
ratingSC='CompRating'
df1C = pandas.read_csv("results/stats-H1-2019-{}.csv".format(ratingSC), names=['X', ratingC, 'H1-2009-' + ratingSC], skiprows=[0])
df1C.head()
```

```python
df2 = pandas.read_csv("results/stats-H2-2019-{}.csv".format(ratingS), names=['X', rating, 'H2-2009-' + ratingS], skiprows=[0])
df2.head()
```

```python
df2C = pandas.read_csv("results/stats-H2-2019-{}.csv".format(ratingSC), names=['X', ratingC, 'H2-2009-' + ratingSC], skiprows=[0])
df2C.head()
```

```python
df3 = pandas.read_csv("results/stats-Year-2019-{}.csv".format(ratingS), names=['X', rating, 'Year-2009-' + ratingS], skiprows=[0])
df3.head()
```

```python
df3C = pandas.read_csv("results/stats-Year-2019-{}.csv".format(ratingSC), names=['X', ratingC, 'Year-2009-' + ratingSC], skiprows=[0])
df3C.head()
```

```python
ax = df1.plot(title="{} Study".format(rating), x=rating, y='H1-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)
ax.set_ylabel("% Change")
ax.set_xlabel(rating)
df2.plot(ax=ax, x=rating, y='H2-2009-' + ratingS, grid=True, legend=True)
df3.plot(ax=ax, x=rating, y='Year-2009-' + ratingS, grid=True, legend=True)
plt.savefig("plot-{}.png".format(ratingS))
```

```python
axC = df1C.plot(title="{} Study".format(ratingC), x=ratingC, y='H1-2009-' + ratingSC, figsize=(15, 9), grid=True, legend=True)
df2C.plot(ax=axC, x=ratingC, y='H2-2009-' + ratingSC, grid=True, legend=True)
df3C.plot(ax=axC, x=ratingC, y='Year-2009-' + ratingSC, grid=True, legend=True)
axC.set_ylabel("% Change")
axC.set_xlabel(ratingC)
plt.savefig("plot-{}.png".format(ratingSC))
```

```python
axEC = df1.plot(title="{}-{} Study".format(rating, ratingS), x=rating, y='H1-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)
df2.plot(ax=axEC, x=rating, y='H2-2009-' + ratingS, grid=True, legend=True)
df3.plot(ax=axEC, x=rating, y='Year-2009-' + ratingS, grid=True, legend=True)

df1C.plot(ax=axEC, x=ratingC, y='H1-2009-' + ratingSC, grid=True, legend=True)
df2C.plot(ax=axEC, x=ratingC, y='H2-2009-' + ratingSC, grid=True, legend=True)
df3C.plot(ax=axEC, x=ratingC, y='Year-2009-' + ratingSC, grid=True, legend=True)

ax.set_ylabel("% Change")
ax.set_xlabel("Rating")
plt.savefig("plot-{}-{}.png".format(ratingS, ratingSC))
```
