```python
import pandas
import matplotlib.pyplot as plt
```

```python
rating='Composite Rating'
ratingS='CompRating'
df1 = pandas.read_csv("results/stats-Q1-2019-{}.csv".format(ratingS), names=['X', rating, 'Q1-2009-' + ratingS], skiprows=[0])
df1.head()
```

```python
df2 = pandas.read_csv("results/stats-H1-2019-{}.csv".format(ratingS), names=['X', rating, 'H1-2009-' + ratingS], skiprows=[0])
df2.head()
```

```python
df3 = pandas.read_csv("results/stats-Year-2019-{}.csv".format(ratingS), names=['X', rating, 'Year-2009-' + ratingS], skiprows=[0])
df3.head()
```

```python
ax = df1.plot(title="{} Study".format(rating), x=rating, y='Q1-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)
ax.set_ylabel("% Change")
ax.set_xlabel(rating)
df2.plot(ax=ax, x=rating, y='H1-2009-' + ratingS, grid=True, legend=True)
df3.plot(ax=ax, x=rating, y='Year-2009-' + ratingS, grid=True, legend=True)
plt.savefig("plot-{}-q1h1y1.png".format(ratingS))
```
