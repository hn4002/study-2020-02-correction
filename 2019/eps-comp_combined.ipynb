```python
import pandas
import matplotlib.pyplot as plt
```

```python
xlabel = "Rating"
rating='EPS Rating'
ratingS='EPSRating'
df1 = pandas.read_csv("results/stats-H1-2019-{}.csv".format(ratingS), names=['X', xlabel, 'H1-2009-' + ratingS], skiprows=[0])
df1.head()
```

```python
ratingC='Composite Rating'
ratingSC='CompRating'
df1C = pandas.read_csv("results/stats-H1-2019-{}.csv".format(ratingSC), names=['X', xlabel, 'H1-2009-' + ratingSC], skiprows=[0])
df1C.head()
```

```python
df2 = pandas.read_csv("results/stats-H2-2019-{}.csv".format(ratingS), names=['X', xlabel, 'H2-2009-' + ratingS], skiprows=[0])
df2.head()
```

```python
df2C = pandas.read_csv("results/stats-H2-2019-{}.csv".format(ratingSC), names=['X', xlabel, 'H2-2009-' + ratingSC], skiprows=[0])
df2C.head()
```

```python
df3 = pandas.read_csv("results/stats-Year-2019-{}.csv".format(ratingS), names=['X', xlabel, 'Year-2009-' + ratingS], skiprows=[0])
df3.head()
```

```python
df3C = pandas.read_csv("results/stats-Year-2019-{}.csv".format(ratingSC), names=['X', xlabel, 'Year-2009-' + ratingSC], skiprows=[0])
df3C.head()
```

```python
ax = df1.plot(title="{} Study".format(rating), x=xlabel, y='H1-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)
ax.set_ylabel("% Change")
ax.set_xlabel(rating)
df2.plot(ax=ax, x=xlabel, y='H2-2009-' + ratingS, grid=True, legend=True)
df3.plot(ax=ax, x=xlabel, y='Year-2009-' + ratingS, grid=True, legend=True)
plt.savefig("plot-{}.png".format(ratingS))
```

```python
axC = df1C.plot(title="{} Study".format(ratingC), x=xlabel, y='H1-2009-' + ratingSC, figsize=(15, 9), grid=True, legend=True)
df2C.plot(ax=axC, x=xlabel, y='H2-2009-' + ratingSC, grid=True, legend=True)
df3C.plot(ax=axC, x=xlabel, y='Year-2009-' + ratingSC, grid=True, legend=True)
axC.set_ylabel("% Change")
axC.set_xlabel(xlabel)
plt.savefig("plot-{}.png".format(ratingSC))
```

```python
axEC = df1.plot(title="H1 / {} - {} Study".format(rating, ratingC), x=xlabel, y='H1-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)

df1C.plot(ax=axEC, x=xlabel, y='H1-2009-' + ratingSC, grid=True, legend=True)

ax.set_ylabel("% Change")
ax.set_xlabel(xlabel)
plt.savefig("plot-{}-{}-h1.png".format(ratingS, ratingSC))
```

```python
axEC = df2.plot(title="H2 / {} - {} Study".format(rating, ratingC), x=xlabel, y='H2-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)

df2C.plot(ax=axEC, x=xlabel, y='H2-2009-' + ratingSC, grid=True, legend=True)

ax.set_ylabel("% Change")
ax.set_xlabel(xlabel)
plt.savefig("plot-{}-{}-h2.png".format(ratingS, ratingSC))
```

```python
axEC = df3.plot(title="Year-2019 / {} - {} Study".format(rating, ratingC), x=xlabel, y='Year-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)

df3C.plot(ax=axEC, x=xlabel, y='Year-2009-' + ratingSC, grid=True, legend=True)

ax.set_ylabel("% Change")
ax.set_xlabel(xlabel)
plt.savefig("plot-{}-{}-Year2019.png".format(ratingS, ratingSC))
```

```python
axEC = df1.plot(title="Q1/H1/Year2019 - {}-{} Study".format(rating, ratingS), x=xlabel, y='H1-2009-' + ratingS, figsize=(15, 9), grid=True, legend=True)
df2.plot(ax=axEC, x=xlabel, y='H2-2009-' + ratingS, grid=True, legend=True)
df3.plot(ax=axEC, x=xlabel, y='Year-2009-' + ratingS, grid=True, legend=True)

df1C.plot(ax=axEC, x=xlabel, y='H1-2009-' + ratingSC, grid=True, legend=True)
df2C.plot(ax=axEC, x=xlabel, y='H2-2009-' + ratingSC, grid=True, legend=True)
df3C.plot(ax=axEC, x=xlabel, y='Year-2009-' + ratingSC, grid=True, legend=True)

ax.set_ylabel("% Change")
ax.set_xlabel("Rating")
plt.savefig("plot-{}-{}-q1h1y1.png".format(ratingS, ratingSC))
```
